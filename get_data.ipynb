```python
from nba_api.stats.static import teams
from nba_api.stats.endpoints import leaguegamefinder

import pandas as pd
import sqlite3
```

```python
team_data=teams.get_teams()
```

```python
connection = sqlite3.connect('data/nba.db')
```

```python
for t in team_data:
    gamefinder = leaguegamefinder.LeagueGameFinder(team_id_nullable=t['id'])
    games = gamefinder.get_data_frames()[0]
    games['opponent']=games.MATCHUP.apply(lambda x: x[-3:])#x.split(' @ '))
    games.to_sql(t['abbreviation'], connection, if_exists='replace')
    
```

```python
pd.read_sql('''
SELECT 
*
FROM sqlite_master 
where type='table'
''', connection)
```

```python
pd.read_sql("select * from sas", connection).columns
```
